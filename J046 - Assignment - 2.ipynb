```python
def layer_size(x,y):
    n_x=x.shape[0]
    n_h=6
    n_y=y.shape[0]
    return(n_x,n_h,n_y)
```

```python
layer_size(x,y)
```

```python
def int_value(n_x,n_h,n_y):
    np.random.seed(1)
    w1=np.random.randn(n_h,n_x)*0.01
    b1=np.zeros(shape=(n_h,1))
    w2=np.random.rand(n_y,n_h)*0.01
    b2=np.zeros(shape=(n_y,1))
    parameters = {"w1":w1,"b1":b1,"w2":w2,"b2":b2}
    
    return parameters
```

```python
n_x,n_h,n_y=layer_size(x,y)
parameters=int_value(n_x,n_h,n_y)
```

```python
parameters
```

```python
def sigmoid(z):
    s=1/(1+np.exp(-z))
    return s
```

```python
def fwd_prg(x,parameters):
    w1=parameters['w1']
    b1=parameters['b1']
    w2=parameters['w2']
    b2=parameters['b2']
    
    z1=np.dot(w1,x)+b1
    a1=np.tanh(z1)
    z2=np.dot(w2,a1)+b2
    a2=sigmoid(z2)
    
    catch={"z1":z1,"a1":a1,"z2":z2,"a2":a2}
    
    return a2, catch
```

```python
a2, catch= fwd_prg(x,parameters)
```

```python
def compute_cost(y,a2,parameters):
    m=y.size
    cost=-np.sum(y*np.log(a2)+(1-y)*np.log(1-a2))/m
    return cost
```

```python
compute_cost(y,a2,parameters)
```
